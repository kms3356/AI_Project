```python
import numpy as np
from math import e
```

```python
# 키를 보고 농구선수인지 예측
X = np.array([160, 170, 180, 190])
y = np.array([0, 0, 1, 1])

X, y
```

```python
# 입력값 정규화
X_mean = np.mean(X)
X_std = np.std(X)

X_norm = (X - X_mean) / X_std
X_mean, X_std, X_norm
```

```python
#  가중치, 편향 초기값 설정
a = 0.1
b = 0.0
```

```python
# sigmoid 함수 정의
def sigmoid(H):
    return 1 / (1 + e ** (-H))

```

```python
# 학습 전 예측 확인
H = a * X_norm + b
z = sigmoid(H)
H, z
```

```python
# Cost 계산. 예측이 얼마나 틀렸는지
# log(0)방지하기 위해 z 범위 제한
epsilon = 1e-7
z_safe = np.clip(z, epsilon, 1 - epsilon)

costs = - (y * np.log(z_safe) + (1 - y) * np.log(1 - z_safe))
mean_cost = np.mean(costs)
costs, mean_cost
```

```python
# dC/da = (z - y) * X_norm
# dC/db = z - y
# grad_a = mean((z-y) * X_norm)
# grad_b = mean(z-y)
epochs = 1000
learning_rate = 0.1
for epoch in range(epochs):
    H = a * X_norm + b
    z = sigmoid(H)
    z_safe = np.clip(z, epsilon, 1 - epsilon)

    grad_a = np.mean((z - y) * X_norm)
    grad_b = np.mean(z - y)

    a -= learning_rate * grad_a
    b -= learning_rate * grad_b

    costs = - (y * np.log(z_safe) + (1 - y) * np.log(1 - z_safe))
    mean_cost = np.mean(costs)

    if epoch % 100 == 0 or epoch == epochs - 1:
        print(f'Epoch {epoch}, 평균 비용(Cost): {mean_cost:.6f}, a: {a}, b: {b}')
```

```python
input_height = 185
input_norm = (input_height - X_mean) / X_std

H_input = a * input_norm + b
probability = sigmoid(H_input)

print(f'\n키가 {input_height}cm인 사람의 농구선수 확률: {probability:.4f}')
if probability >= 0.5:
    print('판별 결과 : 농구선수입니다.')
else:
    print('판별 결과: 농구선수가 아닙니다.')
```
