```python
import numpy as np
from math import e
```

```python
X = np.array([160, 170, 180, 190])
X
```

```python
y = np.array([0, 0, 1, 1])
```

```python
a = 0.1
b = -17
```

```python
H = a * X + b
H
```

```python
def sigmoid(H):
    return 1 / (1 + e**(-H))
```

```python
z = sigmoid(H)
z
```

```python
# 비용(정답을 얼마나 잘맞췄는지에 따른 벌점) 계산. y가 1일때 -y*log(z), y가 0일때 -(1-y)*log(1-z). 낮으면 잘맞춘거임
costs = - (y * np.log(z) + (1 - y) * np.log(1 - z))
costs
```

```python
mean_cost = np.mean(costs)
mean_cost
```
